1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even(lis):
    sum = 0
    for i in lis:
        if i % 2 == 0:
            sum = sum + i
    return sum

lis = [12,23,34,45,56]
sum_of_even(lis)
```

2. Create a Python function that accepts a string and returns the reverse of that string.

```python
def reverse_str(x):
    x = x[::-1]
    return x

name = "koushik"
reverse_str(name)
```

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

```python
def square_of_list(x):
    new_list = []
    for i in x:
        new_list.append(i**2)
    return new_list

lis = [1,2,3,4,5]
square_of_list(lis)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def check_prime(x):
    if x<2:
        return False
    for i in range(2, int(x**0.5)+1):
        if x % i == 0:
            return False
    return True

for num in range(1,201):
    if check_prime(num):
        print(f"{num} is prime")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
def fibonacci(x):
    a, b= 0, 1
    count = 0
    while count<x:
        yield a
        a, b = b, a+b
        count +=1
fib_iter= iter(fibonacci(10))
print(next(fib_iter))
print(next(fib_iter))
print(next(fib_iter))
print(next(fib_iter))
print(next(fib_iter))
print(next(fib_iter))
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(x):
    for i in range(x+1):
        yield i**2
value = power_of_two(15)
print(list(value))
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def line_by_line(x):
    for line in x.split('\n'):
        yield line

file = "In a foggy seaside town, there lived an old clockmaker named Eli. \nHis shop was tiny, tucked between a fish market and a crumbling lighthouse. \nPeople said Eli could fix any clockgrandfather clocks, wristwatches, even cuckoos with missing birds."

gen = line_by_line(file)
print(next(gen))
print(next(gen))
print(next(gen))
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
str1 = "koushik"
vowels = "aeiou"
func = filter(lambda x: x.lower() not in vowels, str1)
result = ''.join(func)
print(result)
```

11. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
order is smaller than 100,00 .

Write a Python program using lambda and map.

```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99]
]

order_list = list(map(lambda x: (x[0], x[2] * x[3] if x[2] * x[3] >= 100 else x[2] * x[3] + 10, x[2]), orders))

print(order_list)
```
